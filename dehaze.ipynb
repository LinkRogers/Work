```python
import numpy as np
import skimage.data
from skimage.measure import compare_psnr
import matplotlib.image as mpimg

img=mpimg.imread('7.jpg')

i,j, k=img.shape
i

def dark_image(img):
    i,j,k=img.shape 
    dark_im = np.zeros((i,j))
    for i_ in range(i):
        for j_ in range(j):
            dark_im[i_,j_] = np.min([img[i_,j_,0],img[i_,j_,1],img[i_,j_,2]])
    return dark_im

dark_im = dark_image(img)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(dark_im)
```
