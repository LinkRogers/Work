```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

from tensorflow.keras import layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Input
from tensorflow.keras.callbacks import TensorBoard
```

```python
import multiprocessing
from multiprocessing import Pool
from functools import partial

def get_data_integer_color(activations,colors):
    f = partial(aplt.assign_color,activations)
    i_color=0
    data_activation_color=[]
    pool = Pool(8)
    
    for y in pool.map(f, range(len(activations))):
        i_color+=1
        data_activation_color.append(y)

    data_activation_color = np.asarray(data_activation_color)
    pool.close()
    pool.join()
    
    return data_activation_color
```

```python
#Prepare data
N=40000
L=4
X = L*np.random.rand(N,2) 
X[:,0] = X[:,0]-L/2
X[:,1] = X[:,1]-L/2

Y = np.zeros([N,2])

for i in range(N):
    if (X[i,0]-0.6)**2+(X[i,1]-0.6)**2<0.64:
        Y[i,0]=1
    else:
        Y[i,1]=1

zero_class = np.where(Y[:, 0])
one_class = np.where(Y[:, 1])

plt.figure(figsize=(8, 8))
plt.scatter(X[zero_class, 0], X[zero_class, 1], s=1, facecolors='b', linewidths=1, label='Class 1')
plt.scatter(X[one_class, 0], X[one_class, 1], s=1, facecolors='r', linewidths=1, label='Class 2')
plt.show()

data = X
labels = Y
```

```python
model = tf.keras.Sequential()
model.add(Dense(3, input_shape=(2,), activation='relu',bias_initializer='random_uniform'))
model.add(Dense(3, activation='relu',bias_initializer='random_uniform'))
model.add(Dense(2, activation='softmax',bias_initializer='random_uniform'))
# Configure a model for categorical classification.
model.compile(optimizer=tf.keras.optimizers.Adam(),
                  loss=tf.keras.losses.categorical_crossentropy,
                  metrics=[tf.keras.metrics.categorical_accuracy])
```

```python
history=model.fit(data, labels, epochs=5,batch_size=int(N/800))
```
